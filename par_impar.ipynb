## Par impar
Crea un algoritmo para determinar si un número es par o impar

## Flujograma diseñado
![./img/glujograma.png](./img/flujograma.png)

```python
# pedir texto al usuario
texto_usuario = input("Ingresa un número")

# validar si el texto es un número
# si no es un número volver a perdir texto
while not texto_usuario.isdigit():
    texto_usuario = input("Ingresa un número")

# Convertir texto a entero
num = int(texto_usuario)
print(num)
```

```python
# División entera entre 2
division = num // 2
print(division)
```

```python
# Multiplicar por 2
multiplicacion = division * 2
print(multiplicacion)
```

```python
# Comparar numero ingresado con multiplicación

# Si al multiplicar por 2 se obtiene el mismo resultado
# Significa que el número es par

if multiplicacion == num:
    print(f"El número {num} es par")
else:
    print(f"El número {num} es impar")
```
